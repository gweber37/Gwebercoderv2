## **Resumen**

Este proyecto tiene como objetivo analizar un conjunto de datos sobre el comportamiento de los jugadores en línea. El dataset incluye variables demográficas y métricas relacionadas con la participación y compromiso en juegos en línea. A través de este análisis, buscamos identificar patrones y tendencias que puedan ayudar a mejorar la retención y el compromiso de los jugadores. Las preguntas de interés incluyen la correlación entre la edad y las horas jugadas, la influencia del género en la selección del tipo de juego, y cómo la edad afecta el nivel de compromiso de los jugadores. Utilizaremos pandas para la manipulación de datos, y Matplotlib y Seaborn para crear visualizaciones que nos permitan obtener insights relevantes.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
url='https://github.com/gweber37/Gwebercoderv2/blob/main/online_gaming_behavior_dataset.csv'
df_juegos= pd.read_csv('/content/online_gaming_behavior_dataset.csv')
```

## **Preguntas de intéres o hipótesis**

#1- ¿Existe alguna correlación entre la edad de lo jugadores y la cantidad de horas jugadas?

# Hipótesis: Los jugadores de ciertas edades tienen predisposición a invertir más tiempo en el juego. Por ejemplo las personas más jóvenes podrían tener más tiempo libre para jugar.

# 2-¿Influye la edad  o el género al momento de seleccionar el tipo de juego?

# Hipotesis: Los jugadores de algún genero podria tener predominancia al momento de Jugar on line . Por ejemplo , los hombres independientemente de la edad probablemente juegen más y más tiempo que las mujeres.

# 3-¿Es la edad un factor importante respecto al ¨engagement¨o compromiso al juego?

# Hipótesis: Los individuos más jóvenes podrian sentirse más comprometidos que los más grandes.

### 1. Distribución de Edades de los Jugadores

```python
plt.figure(figsize=(10, 6))
```

```python
sns.histplot(df_juegos['Age'], bins=30, kde=True)
plt.title('Distribución de Edades de los Jugadores')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
```

**Interpretación:** Este gráfico ilustra la distribución de edades de los jugadores en el conjunto de datos. Podemos observar la tendencia general de las edades de los jugadores y detectar posibles grupos de edad predominantes.


### 2. Tipos de Juego por Género

```python
plt.figure(figsize=(10, 6))
```

```python
sns.countplot(x='GameGenre', hue='Gender', data=df_juegos)
plt.title('Tipos de Juego por Género')
plt.xlabel('Tipo de Juego')
plt.ylabel('Cantidad')
```

**Interpretación:** Este gráfico nos permite visualizar si existe una preferencia de tipo de juego según el género de los jugadores. Podemos identificar si ciertos géneros de juegos son más populares entre hombres o mujeres.


### 3. Horas Jugadas por Edad

```python
plt.figure(figsize=(10, 6))
```

```python
sns.scatterplot(x='Age', y='PlayTimeHours', data=df_juegos)
plt.title('Horas Jugadas por Edad')
plt.xlabel('Edad')
plt.ylabel('Horas Jugadas')
```

### 4. Nivel de Compromiso por Edad y Género

```python
plt.figure(figsize=(10, 6))
```

```python
sns.barplot(x='EngagementLevel', y='InGamePurchases', hue='Gender', data=df_juegos)
plt.title('Compras en el Juego por Nivel de Compromiso')
plt.xlabel('Nivel de Compromiso')
plt.ylabel('Compras en el Juego')
```

**Interpretación:** Este gráfico nos permitirá comprender si existe alguna correlación entre la edad de los jugadores y la cantidad de horas que dedican al juego. Podremos confirmar o refutar nuestra hipótesis sobre la relación entre la edad y el tiempo de juego.



```python
plt.figure(figsize=(10, 6))
```

```python
sns.scatterplot(x='SessionsPerWeek', y='AvgSessionDurationMinutes', hue='Gender', data=df_juegos)
plt.title('Sesiones por Semana y Duración Promedio de las Sesiones')
plt.xlabel('Sesiones por Semana')
plt.ylabel('Duración Promedio de las Sesiones (minutos)')
```


**Interpretación:** Este gráfico nos permitirá comprender cómo se relacionan las sesiones por semana con la duración promedio de las mismas y si existen diferencias según el género de los jugadores.



```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='SessionsPerWeek', y='AvgSessionDurationMinutes', hue='Gender', data=df_juegos)
plt.title('Sesiones por Semana y Duración Promedio de las Sesiones')
plt.xlabel('Sesiones por Semana')
plt.ylabel('Duración Promedio de las Sesiones (minutos)')

```

**Interpretación:** Este gráfico nos permitirá comprender la relación entre las sesiones por semana y la duración promedio de las mismas, y si existen diferencias según el género de los jugadores.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='EngagementLevel', y='InGamePurchases', hue='Gender', data=df_juegos)
plt.title('Compras en el Juego por Nivel de Compromiso')
plt.xlabel('Nivel de Compromiso')
plt.ylabel('Compras en el Juego')
```

**Interpretación:** Este gráfico nos permitirá observar si existe una relación entre el nivel de compromiso de los jugadores y las compras que realizan en el juego, así como posibles diferencias entre géneros.
